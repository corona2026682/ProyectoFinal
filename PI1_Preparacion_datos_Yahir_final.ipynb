Algunos campos, como belongs_to_collection, production_companies y otros, están anidados ¡deberán desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

```python
import ast
def parse_dict_string(value):
    if pd.isna(value):
        return None
    else:
        return ast.literal_eval(value)
    
```

```python
# Para aplicar parse_dict_string para pasar de diccionario a string
data["belongs_to_collection"] = data["belongs_to_collection"].apply(parse_dict_string)

# Función para extraer el valor de "name" de los registros en la columna "collection"
def extract_name(record):
    if record is not None:
        if isinstance(record, dict):
            return record.get('name')
    else:
        return None

# Aplicar la función a la columna "collection" y crear una nueva columna "name_to_collection"
data['name_to_collection'] = data['belongs_to_collection'].apply(extract_name)
```

```python
# Para aplicar parse_dict_string para pasar de diccionario a string
data['genres'] = data['genres'].apply(parse_dict_string)

# Función para extraer los valores de "name" de cada registro
def extract_name(record):
    return [item['name'] for item in record]

# Aplicar la función a la columna 'genres' y expandir los resultados en una nueva columna 'name_genres'
data['name_genres'] = data['genres'].apply(lambda x: extract_name(x))
```

```python
# Para ver la nueva columna "name_genres"
data.head(1)
```

```python
def desanida_name(columna, lugar_name_int): # en lugar_name_int hay que colocar en que posicion del diccionario esta "name"
    global data
    # Funcion para convertir de diccionario a string
    import ast
    def parse_dict_string(value):
        if pd.isna(value):
            return None
        else:
            return ast.literal_eval(value)
        
    # Aplicamos la funcion para pasar de diccionario a string y desanidamos
    data[columna] = data[columna].apply(parse_dict_string)
    df_desanidado = pd.json_normalize(data[columna]).fillna("")
    #return df_desanidado.head()
    
    # Obtener la cantidad de columnas en el DataFrame
    cantidad_columnas = len(df_desanidado.columns)

    # Creamos un nuevo dataframe para concatenarlo luego de desanidar completamente el dataframe anterior (df_desanidado)
    concatenar_data = pd.DataFrame()

    # Iterar sobre las columnas del DataFrame
    for i in range(cantidad_columnas):
        df_desanidado_2 = pd.json_normalize(df_desanidado[i]).fillna("")
        # Concatenar los dataframes
        concatenar_data = pd.concat([concatenar_data, df_desanidado_2], axis=1)
    
    # Función para juntar los valores en una lista
    def juntar_valores_en_lista(row):
        lista_valores = []
        for valor in row:
            if valor.strip():  # Verificar si el valor no está vacío después de eliminar espacios en blanco
                lista_valores.append(valor)
        return lista_valores

    # Aplicar la función a las columnas y crear una nueva columna 'Lista'
    concatenar_data['name'] = concatenar_data["name"].apply(juntar_valores_en_lista, axis=1)

    # Selección de columnas por número de columna
    df_seleccionado = concatenar_data.iloc[:, lugar_name_int]
    df_seleccionado = pd.DataFrame(df_seleccionado)

    # Concatenamos df_desanidado["name"] a data
    data = pd.concat([data, df_seleccionado['name']], axis=1)
```

Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0.

```python
data['revenue'] = data['revenue'].fillna(0)
data['budget'] = data['budget'].fillna(0)
```

Los valores nulos del campo release date deben eliminarse.

```python
data = data.dropna(subset=['release_date'])
data.shape
```

De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year donde extraerán el año de la fecha de estreno.

```python
# Convertir la columna 'release_date' al formato de fecha
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Eliminar las filas con valores no válidos en la columna 'release_date'
data = data.dropna(subset=['release_date'])
```

```python
data.info()
```

```python
# Crear una nueva columna 'release_year' con el año de la columna 'release_date'
data['release_year'] = data['release_date'].dt.year
data.head(1)
```

Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget, dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, deberá tomar el valor 0.

```python
# Convertir la columna 'budget' a tipo float
data['budget'] = data['budget'].astype(float)

# Calcular la columna "return" con la división de "revenue" entre "budget", teniendo en cuenta los casos nulos y 0
data['return'] = data.apply(lambda row: row['revenue'] / row['budget'] if row['revenue'] and row['budget'] else 0, axis=1)
```

```python
# Aplicamos la funcion para pasar de diccionario a string para luego poder desanidar
credits["cast"] = credits["cast"].apply(parse_dict_string)
```

```python
# Función para extraer los valores deseados de cada registro
def extract_values(registros):
    names = [registro['name'] for registro in registros]
    return pd.Series({'names': names})

# Aplicar la función a la columna 'registros' y expandir los resultados en nuevas columnas
credits['names_actors'] = credits['cast'].apply(extract_values)
```

```python
# Aplicamos la funcion para pasar de diccionario a string para luego poder desanidar
credits["crew"] = credits["crew"].apply(parse_dict_string)
```

```python
# Función para extraer el valor de "name" de los registros con "job" igual a "Director"
def extraer_nombre_director(registros):
    for registro in registros:
        if registro['job'] == 'Director':
            return registro['name']
    return None

# Aplicar la función a la columna y crear una nueva columna llamada "name_director"
credits['name_director'] = credits['crew'].apply(lambda x: extraer_nombre_director(x))
```

```python
# Crearemos un dataframe para concatenar solo con las columnas que necesitamos
# Columnas a seleccionar
columnas_seleccionadas = ["id", "names_actors", "name_director"]

# Crear nuevo DataFrame con las columnas seleccionadas
credits_concatenar = credits.loc[:, columnas_seleccionadas]
```

Preparamos la data para utilizarla en el modelo de machine learning


```python
# Seleccionamos las columnas
data_preparada_ML = data[["name_genres", "title", "release_date", "release_year", "popularity", "vote_count", "vote_average", "names_actors", "name_director", "budget", "revenue", "return"]]

# Filtrar las películas desde el año 1995 hasta el año 2020 para mermar la cantidad de datos
data_preparada_ML = data_preparada_ML[(data_preparada_ML['release_year'] >= 1995) & (data_preparada_ML['release_year'] <= 2020)]

# Filtrar las peliculas que tengan vote_average mayor a 0 para mermar la cantidad de datos
data_preparada_ML = data_preparada_ML[data_preparada_ML['vote_average'] > 0]

# Eliminamos duplicados de la columna title porque me sale un error en render si hay dos peliculas con el mismo nombre
data_preparada_ML = data_preparada_ML.drop_duplicates(subset="title")

# Tomamos una parte del dataset como ejemplo ya que en render se puede utilizar hasta 512mb de RAM y si ponemos todos los registros supera el limite de RAM de render.
data_preparada_ML = data_preparada_ML.sample(8000, random_state=84)

# Creamos el archivo csv
data_preparada_ML.to_csv("data_preparada_ML.csv", index=False)
```
